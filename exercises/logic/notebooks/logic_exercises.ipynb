## Exercise 6 – Resolution 🐍

**Resolution** is a proof technique for propositional logic. It works by refutation:
to prove that $KB \models \phi$, we add $\lnot\phi$ to the KB and try to derive a **contradiction** (the empty clause $\{\}$).

### Clauses and literals

In resolution, every formula must be in **CNF** — a conjunction of **clauses**, where each clause is a **disjunction of literals**.

We represent a clause as a Python **set of strings**:
```python
{'P', '¬Q', 'R'}   # means  P ∨ ¬Q ∨ R
```

### The resolution rule

Given two clauses that contain **complementary literals** (e.g. `P` and `¬P`),  
we can derive a new clause — the **resolvent** — by combining both clauses minus those complementary literals:

```
C1 = {P, A, B}     C2 = {¬P, C, D}
resolvent = {A, B, C, D}   (remove P and ¬P, merge the rest)
```

If the resolvent is the **empty set** `{}`, we found a contradiction → proof complete!

````python
def negate(literal):
    """
    Returns the negation of a literal.
    negate('P')  -> '¬P'
    negate('¬P') -> 'P'
    """
    if literal.startswith('¬'):
        return literal[1:]          # strip the ¬
    else:
        return '¬' + literal        # prepend ¬


def resolve(c1, c2):
    """
    Try to resolve two clauses (given as sets of literal strings).
    Returns the resolvent set if resolution is possible, else None.

    Example:
        resolve({'P', 'A'}, {'¬P', 'B'})  ->  {'A', 'B'}
        resolve({'P', 'Q'}, {'A', 'B'})   ->  None  (no complementary literals)
    """
    for lit in c1:
        if negate(lit) in c2:
            # YOUR CODE HERE:
            # Build the resolvent:
            #   - take everything from c1 EXCEPT lit
            #   - union with everything from c2 EXCEPT negate(lit)
            # Hint: set difference:  c1 - {lit}
            # Hint: set union:       set_a | set_b
            resolvent = None  # replace this
            return resolvent
    return None   # no complementary literals found


# ── Quick test ──────────────────────────────────────────────────────────────
r = resolve({'P', 'A'}, {'¬P', 'B'})
print(f"resolve({{'P','A'}}, {{'¬P','B'}}) = {r}")
print(f"Expected: {{'A', 'B'}}")
print()
r2 = resolve({'P', 'Q'}, {'A', 'B'})
print(f"resolve({{'P','Q'}}, {{'A','B'}}) = {r2}")
print(f"Expected: None")
````

````python
def pl_resolution(kb_clauses, goal_literals):
    """
    Attempt to prove a goal from KB using resolution by refutation.

    kb_clauses    : list of clauses (each clause is a set of literal strings)
    goal_literals : set of literals representing the goal (e.g. {'Magical'})

    How it works:
      1. Negate every literal in goal_literals and add each as a unit clause.
      2. Repeatedly try to resolve pairs of clauses.
      3. If we derive the empty clause {} → contradiction → goal is PROVED.
      4. If no new clauses can be derived → goal CANNOT be proved.
    """
    # Step 1: negate the goal and add as unit clauses
    negated_goal = [frozenset({negate(lit)}) for lit in goal_literals]

    # Work with frozensets so we can put clauses in a set
    clauses = set(frozenset(c) for c in kb_clauses) | set(negated_goal)
    new_clauses = set()

    iteration = 0
    while True:
        iteration += 1
        clause_list = list(clauses)

        # Try every pair of clauses
        for i in range(len(clause_list)):
            for j in range(i + 1, len(clause_list)):
                c1 = set(clause_list[i])
                c2 = set(clause_list[j])

                # YOUR CODE HERE:
                # 1. Call resolve(c1, c2) to get the resolvent
                # 2. If the resolvent is the EMPTY SET → return True (proved!)
                # 3. Otherwise if resolvent is not None → add it to new_clauses
                pass

        # If no new clauses were found, we cannot prove the goal
        if new_clauses.issubset(clauses):
            return False

        clauses |= new_clauses


print("pl_resolution defined ✅")
````

## Exercise 7 – First-Order Logic: Formula Evaluator 🐍

FOL formulas quantify over a **domain** of objects.  
We can check FOL formulas computationally by iterating over a finite domain.

We will use a small domain of animals: cats and dogs, big and small.

A **model** in FOL is:
- A domain (set of objects)
- An interpretation of each predicate (which objects satisfy it)

```python
domain = ['cat1', 'cat2', 'dog1']
big    = {'cat1', 'dog1'}      # big(x) is true for these objects
cat    = {'cat1', 'cat2'}      # cat(x) is true for these objects
```

````python
# ── Domain setup ─────────────────────────────────────────────────────────────
domain = ['cat1', 'cat2', 'dog1', 'dog2']
big    = {'cat1', 'dog1', 'dog2'}   # big(x)
cat    = {'cat1', 'cat2'}           # cat(x)


# ── Helper: check a formula for all x in domain ──────────────────────────────

def forall(domain, predicate):
    """Returns True if predicate(x) is True for EVERY x in domain."""
    return all(predicate(x) for x in domain)

def exists(domain, predicate):
    """Returns True if predicate(x) is True for AT LEAST ONE x in domain."""
    return any(predicate(x) for x in domain)


# ── 7.1: Evaluate the 7 formulas from the worksheet ─────────────────────────
# Formula 1: ∃x. (big(x) ∧ cat(x))  →  "There exists a big cat"
f1 = exists(domain, lambda x: x in big and x in cat)
print(f"Formula 1 — ∃x.(big∧cat):        {f1}   ← expected True (cat1 is big and a cat)")

# Formula 4: ∀x. (big(x) ∧ cat(x))  →  "All things are big cats"
f4 = forall(domain, lambda x: x in big and x in cat)
print(f"Formula 4 — ∀x.(big∧cat):        {f4}  ← expected False (dog2 is not a cat)")

# Formula 5: ∀x. (big(x) → cat(x))  →  "All big things are cats"
f5 = forall(domain, lambda x: x not in big or x in cat)
print(f"Formula 5 — ∀x.(big→cat):        {f5}  ← expected False (dog1 is big but not a cat)")

print()

# YOUR CODE HERE — evaluate formulas 2, 3, 6, 7:
# Formula 2: ∃x. (big(x) → cat(x))   (remember: A→B = ¬A ∨ B)
# Formula 3: ∃x. (big(x) ∨ cat(x))
# Formula 6: ∀x. (big(x) ∨ cat(x))
# Formula 7: ∀x. (big(x) ∨ ¬cat(x))
pass
````
